```python
def stats_missing_data(df: pd.DataFrame) -> None:
    stats_missing_data = []
    total = df.shape[0] 
    
    for col in df.columns:
        if df[col].isna().any():
            amount = df[col].isna().sum()  
            dict_missing_data = { 
                col: {
                    'amount': amount,
                    'percentage': round(100 * amount / total, 2)
                }
            }
            stats_missing_data.append(dict_missing_data)
    
    for stat in stats_missing_data:
        print(stat)
```

```python
stats_missing_data(df=df)
```
